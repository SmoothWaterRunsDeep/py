当你写 dataset = CIFAR10(..., transform=ToTensor()) 时，每当你通过 dataset[idx] 取数据，PyTorch 就会自动在后台执行一次转换，把原始图像变成 Tensor。下面一步一步拆开告诉你到底发生了什么。
当你之后写：img, label = train_dataset[5]
Python 会调用：train_dataset.__getitem__(5)
在 CIFAR10 的内部实现里（简化版）：
def __getitem__(self, index):
    img = PIL.Image.open(...)
    
    if self.transform is not None:
        img = self.transform(img)  # ← 就是你传进来的 transform
    
    return img, label
